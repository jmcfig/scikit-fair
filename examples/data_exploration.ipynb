# Dataset Exploration

Basic exploratory data analysis for the three datasets available in scikit-fair: **Adult**, **German Credit**, and **Heart Disease**.

```python
import pandas as pd
from skfair.datasets import load_adult, load_german, load_heart_disease
```

## 1. Adult Dataset

```python
adult = load_adult(preprocessed=False)
df_adult = adult.frame
print(f"Shape: {df_adult.shape}")
df_adult.head()
```

```python
df_adult.race.unique()
```

```python
df_adult.dtypes
```

```python
# Missing values
missing_adult = df_adult.isnull().sum()
missing_adult[missing_adult > 0] if missing_adult.any() else print("No missing values")
```

```python
# Check for '?' placeholders (common in Adult dataset)
for col in df_adult.select_dtypes(include="object").columns:
    n_q = (df_adult[col] == "?").sum()
    if n_q > 0:
        print(f"{col}: {n_q} '?' values ({n_q / len(df_adult) * 100:.1f}%)")
```

```python
df_adult.describe()
```

```python
df_adult.describe(include="object")
```

```python
# Target distribution
df_adult["Probability"].value_counts(normalize=True)
```

```python
# Sensitive attribute distribution
df_adult["sex"].value_counts(normalize=True)
```

```python
# Target by sensitive attribute
pd.crosstab(df_adult["sex"], df_adult["Probability"], normalize="index")
```

## 4. Preprocessed Data

Quick look at what the `preprocessed=True` flag produces.

```python
X_adult, y_adult = load_adult(return_X_y=True, preprocessed=True)
print(f"Adult preprocessed shape: {X_adult.shape}")
print(f"All numeric: {X_adult.select_dtypes(include='number').shape[1] == X_adult.shape[1]}")
X_adult.head()
```

```python
X_german, y_german = load_german(return_X_y=True, preprocessed=True)
print(f"German preprocessed shape: {X_german.shape}")
print(f"All numeric: {X_german.select_dtypes(include='number').shape[1] == X_german.shape[1]}")
X_german.head()
```

```python
X_heart, y_heart = load_heart_disease(return_X_y=True, preprocessed=True)
print(f"Heart preprocessed shape: {X_heart.shape}")
print(f"All numeric: {X_heart.select_dtypes(include='number').shape[1] == X_heart.shape[1]}")
X_heart.head()
```
