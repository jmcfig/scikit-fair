# 01 — Datasets

scikit-fair for now ships five fairness benchmark datasets. Each loader returns a pandas
DataFrame by default and accepts a **preprocessed** flag that controls whether
categorical features are encoded and numeric features are standardised.

## Available datasets

| Loader | Domain | Sensitive attribute | Target |
|---|---|---|---|
| `load_adult` | Income | sex | Income >50K |
| `load_compas` | Recidivism | sex / race | Two-year recidivism |
| `load_german` | Credit | sex | Credit risk |
| `load_heart_disease` | Health | sex | Heart disease |
| `load_ricci` | Employment | Race | Combined test score |

## Raw vs preprocessed — Ricci dataset

With `preprocessed=False` you get the data as-is.

```python
from skfair.datasets import load_ricci

X_raw, y_raw = load_ricci(preprocessed=False)
print("Shape:", X_raw.shape)
X_raw.head()
```

```python
X_raw.dtypes
```

With `preprocessed=True` (the default), categorical columns like Race and
Position are label-encoded

```python
X_pre, y_pre = load_ricci(preprocessed=True)
print("Shape:", X_pre.shape)
X_pre.head()
```

## Raw vs preprocessed — Adult dataset

The Adult dataset has many more categorical features, so the difference is
larger. Preprocessing one-hot encodes categoricals, standardises numerics,
and keeps the sensitive column as-is.

```python
from skfair.datasets import load_adult

X_raw_a, y_raw_a = load_adult(preprocessed=False)
print("Shape:", X_raw_a.shape)
X_raw_a.head()
```

```python
X_pre_a, y_pre_a = load_adult(preprocessed=True)
print("Shape:", X_pre_a.shape)
X_pre_a.head()
```

## Loader API

All five loaders share the same interface:

- `return_X_y` — return `(X, y)` tuple (default `True`); set to `False` for a Bunch object
- `as_frame` — return pandas objects (default `True`); set to `False` for numpy arrays
- `preprocessed` — encode and standardise (default `True`)
- `target_column` — name of the target column (dataset-specific default)
